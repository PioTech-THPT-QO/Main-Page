## Bài 1: Fibonacci Prime
Dãy số Fibonacci là dãy vô hạn các số bắt đầu bằng 1 và 1 (hoặc 0 và 1), các phần tử sau luôn bằng tổng hai phần tử trước nó.

Công thức truy hồi của dãy Fibonacci là:  
![Ảnh công thức dãy Fibonacci](https://wikimedia.org/api/rest_v1/media/math/render/svg/6693d2c78bc8132bb9b65be861148ca574a738ef)  

Số Fibonacci nguyên tố là số thuộc dãy Fibonacci và chỉ có hai ước là 1 và chính nó. VD: 2, 3, 5, 13 là các số Fibonacci nguyên tố.

Cho một số nguyên dương *N*, hãy kiểm tra xem liệu *N* có phải là số **Fibonacci nguyên tố** hay không.

### Mô tả đầu vào:  

- Đọc từ tệp **FBPR.INP** một số nguyên dương duy nhất là *N*. (0 ≤ *N* ≤ 10^9)

### Mô tả đầu ra:  

- Ghi ra tệp **FBPR.OUT** một số duy nhất là **1** nếu *N* nhập là số Fibonacci nguyên tố, ngược lại, ghi **0**.

### Rằng buộc:
- Có 40 % số test *N* ≤ 10^6.
- Có 40 % số test 10^6 ≤ *N* ≤ 10^12.
- Có 20 % số test *N* > 10^15.

| FBPR.INP | FBPR.OUT|
|:--------|:---------|
| 9 | 0 |
| 165580141 | 1 |

## Bài 3: Nhện Con Tìm Đường

Nhà Nam có một lưới ô vuông *N* x *M* (số hàng *M* < *N* số cột). Một chú nhện cần leo từ điểm A lên điểm B (vị trí hai điểm như hình vẽ)(hình vẽ minh hoạ M và N). Mỗi lần leo, chú nhện chỉ có thể bò men theo các **cạnh** của ô vuông: lên đúng một ô hoặc sang phải (**có thể nhiều ô**). Hỏi có bao nhiêu cách để chú nhện có thể từ A lên được B?

![Ảnh Nam](https://i.ibb.co/bvJNjyS/nam.png)

### Mô tả đầu vào:  

- Đọc từ tệp **SPIDER.INP** một dòng duy nhất chứa hai số nguyên dương là *M* và *N* cách nhau bởi dấu cách. (0 ≤ *M* < *N* ≤ 10^3)

### Mô tả đầu ra:  

- Ghi ra tệp **SPIDER.OUT** một số nguyên dương duy nhất là số cách mà chú nhện có thể leo từ điểm A (góc dưới bên trái) đến điểm B (góc trên bên phải).

### Rằng buộc:
- Có 40 % số test 0 < *M* < *N* ≤ 10.
- Có 60 % số test 10 ≤ *M* <> *N* ≤ 100.

| SPIDER.INP | SPIDER.OUT| Giải thích |
|:--------|:---------| :---: |
| 2 3 | 6 | ![giải thích nhện](https://i.ibb.co/tDM8qGR/ne.png) |
| 10 11 | 66 | (Tương tự ví dụ trên ta tìm được 66 cách đi.) |

## Bài 4: Bước Chân Nhiệm Màu

Nhà Dương có cái cầu thang gồm *N* bậc và mỗi bậc được đánh số như hình vẽ.

![Ảnh cách đánh số bậc thang](https://i.ibb.co/TTG1bkW/image.png)

Dương đang đứng ở chân cầu thang và muốn trèo lên đỉnh cầu thang. Số bước cầu thang Dương có thể thực hiện mỗi lần là **một hoặc hai bước**. Hãy giúp Dương tính số cách để leo lên đỉnh cầu thang.

### Mô tả đầu vào:  

- Đọc từ tệp **FLFW.INP** một số nguyên dương duy nhất là *N*. (*N* ≤ 10^3)

### Mô tả đầu ra:  

- Ghi ra tệp **FLFW.OUT** một số duy nhất là số cách để leo lên đỉnh cầu thang.

### Rằng buộc:

- Có 40 % số test *N* < 10^2.
- Có 60 % số test 10^2 < *N* < 10^3.

| FLFW.INP | FLFW.OUT | Giải thích |
|:--------|:---------| :-----------|
| 2 | 2 | Có hai cách để leo lên bậc thứ 2: <ul> <li> 0→1→2 </li> <li> 0→2 </li> </ul> |
| 4 | 5 | Có năm cách để leo lên bậc thứ 4: <ul> <li> 0→1→2→3→4 </li> <li> 0→1→2→4 </li> <li> 0→1→3→4 </li> <li> 0→2→3→4 </li><li> 0→2→4 </li></ul>|
