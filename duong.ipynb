## Bài 1: Số Fibonacci
Dãy số Fibonacci là dãy vô hạn các số bắt đầu bằng 1 và 1 (hoặc 0 và 1), các phần tử sau luôn bằng tổng hai phần tử trước nó.

Công thức truy hồi của dãy Fibonacci là:  
![Ảnh công thức dãy Fibonacci](https://wikimedia.org/api/rest_v1/media/math/render/svg/6693d2c78bc8132bb9b65be861148ca574a738ef)  

Cho một số nguyên dương *N*, hãy kiểm tra xem liệu N có **thuộc** dãy Fibonacci hay không.

### Mô tả đầu vào:  
- Đọc từ tệp **fibo.inp** một số nguyên dương duy nhất là *N*. (0 $\le$ *N* $\le$ 10^9)
### Mô tả đầu ra:  
- Ghi ra tệp **fibo.out** một số duy nhất là **1** nếu *N* nhập vào thuộc dãy Fibonacci, ngược lại, ghi **0**.
### Rằng buộc:
- Có 40 % số test *N* < 10^6.
- Có 40 % số test 10^6 < *N* < 10^12.
- Có 20 % số test N > 10^15.

| fibo.inp | fibo.out|
|:--------|:---------|
| 9 | 0 |
| 165580141 | 1 |
